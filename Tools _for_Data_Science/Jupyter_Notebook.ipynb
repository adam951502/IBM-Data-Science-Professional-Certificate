### Plot the bar chart and pie chart of the programming languages used in the class.

```python
import matplotlib.pyplot as plt
# data
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [8,30,55,100,3]
Percentage = []
for i in students:
    Percentage.append(round(100*i/sum(students),2))

# function to add value labels
def addlabels(x,y):
    for i in range(len(x)):
        axs[0].text(i,y[i],y[i], ha = 'center',  va='bottom')    
    
# bar chart - create first subplot on the left (1 row, 2 columns, position 1)
fig, axs = plt.subplots(1, 2, figsize=(8, 5))
axs[0].bar(langs,students, width=0.3)
axs[0].set_xlabel('Programming Languages')
axs[0].set_ylabel('Number of students')
axs[0].set_title('Programming language used in the class')
addlabels(langs, students)

# pie chart - create second subplot on the right (1 row, 2 columns, position 2)
axs[1].pie(Percentage,labels=langs, autopct='%1.1f%%', radius=1.5)
# axs[1].set_title('Procentage of Programming language used in the class', )
plt.show()
```
